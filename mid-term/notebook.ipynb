## Data Cleaning

```python
df.isnull().sum()
```

```python
df['usd pledged'] = df['usd pledged'].fillna(df['usd pledged'].mean())

print(df.isnull().sum())
```

```python
df['deadline'] = pd.to_datetime(df['deadline'])
df['launched'] = pd.to_datetime(df['launched'])
```

```python
df['state'].value_counts().plot(kind='bar')
```

## Feature Selection and Engineering

```python
df['campaign_duration'] = (df['deadline'] - df['launched']).dt.days

df['launched_dayofweek'] = df['launched'].dt.dayofweek
df['launched_month'] = df['launched'].dt.month

df[['deadline', 'launched', 'launched_dayofweek','launched_month','campaign_duration']].head()
```

```python
df.info()
```

```python
numerical_features = [ 'backers', 'usd_pledged_real','usd_goal_real','launched_month','campaign_duration']
```

```python
categorical_features = ['main_category', 'currency', 'country']
```

```python
print("Numerical Features:")
print(numerical_features)
print("\nCategorical Features:")
print(categorical_features)
```

## Split data

```python
from sklearn.model_selection import train_test_split
df_train_full, df_test = train_test_split(df, test_size=0.2, random_state=42,stratify= df['state'])
df_train, df_val = train_test_split(df_train_full, test_size=0.25, random_state=42)

df_train = df_train.reset_index(drop=True)
df_val = df_val.reset_index(drop=True)
df_test = df_test.reset_index(drop=True)

y_train = df_train.state.values
y_val = df_val.state.values
y_test = df_test.state.values

del df_train['state']
del df_val['state']
del df_test['state']
```

## Train model

```python
from sklearn.feature_extraction import DictVectorizer
from sklearn.ensemble import RandomForestClassifier

train_dict = df_train[categorical_features + numerical_features].to_dict(orient='records')

dv = DictVectorizer(sparse=False)
dv.fit(train_dict)

X_train = dv.transform(train_dict)

model = RandomForestClassifier(n_estimators=92, random_state=42)  # Adjust hyperparameters
model.fit(X_train, y_train)
```

## Training report

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_train)
accuracy = accuracy_score(y_train, y_pred)
print(f"Accuracy: {accuracy}")
print(classification_report(y_train, y_pred))
```

## Test report

```python

test_dict = df_test[categorical_features + numerical_features].to_dict(orient='records')

X_test = dv.transform(test_dict)

y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy: {accuracy}")
print(classification_report(y_test, y_pred))
```

## Let's try to improve it

```python
from sklearn.utils.class_weight import compute_class_weight

class_weights = compute_class_weight(class_weight='balanced', classes=df['state'].unique(), y=df['state'])
class_weights_dict = dict(zip(df['state'].unique(), class_weights))
```

```python
class_weights_dict
```

```python
val_dict = df_val[categorical_features + numerical_features].to_dict(orient='records')


X_val = dv.transform(val_dict)
```

```python
from sklearn.model_selection import GridSearchCV
param_grid = {
    'n_estimators': [90 , 100, 150],
    'max_depth': [None, 10, 20],
    'class_weight': ['balanced', class_weights_dict, None],
}


grid_search = GridSearchCV(estimator=model, param_grid=param_grid, cv=4, scoring='accuracy') # you can use other scoring metrics too, like f1_macro for unbalanced data.
grid_search.fit(X_val, y_val)

best_model = grid_search.best_estimator_
```
